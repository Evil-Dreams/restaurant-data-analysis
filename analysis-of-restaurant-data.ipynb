```python
import pandas as pd
```

```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
data1 = kagglehub.dataset_load(
  KaggleDatasetAdapter.PANDAS,
  "vibishananandhan/restaurant-data-analysis",
  "Restaurant1.csv")
```

```python
data1.head(3)
```

```python
data2 = kagglehub.dataset_load(
  KaggleDatasetAdapter.PANDAS,
  "vibishananandhan/restaurant-data-analysis",
  "Restaurant2.csv")
```

```python
data2.head(3)
```

```python
data=data1.merge(data2,how='outer')
```

```python
data['Invoice No.'].unique()
```

```python
norder=data['Invoice No.'].nunique()
print("Total number of orders: ",norder)
```

```python
total=data['Final Total'].sum()
print("Total Cost collected: ",total)
```

```python
avg=data['Final Total'].sum()/data['Invoice No.'].nunique()
print("Average cost collected from each order: ",avg)
```

```python
data['Date'] = pd.to_datetime(data['Date'])
recent_dates = data['Date'].nlargest(2).unique()
recent_dates = data['Date'].drop_duplicates().nlargest(2).tolist()
filtered_data = data[data['Date'].isin(recent_dates)]
total_revenue = filtered_data['Final Total'].sum()
print(f"Total revenue for the two most recent days: {total_revenue:.2f}")
```

```python
invoice_total = data.groupby('Invoice No.')['Final Total'].sum().loc[28327]
print("Total amount for Invoice No. 28327: ",invoice_total)
```

```python
items_by_qty = data.groupby('Item Name')['Qty.'].sum()
items_by_qty.nlargest(3)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,6))
plt.bar(items_by_qty.nlargest(5).index,items_by_qty.nlargest(5).values)
plt.xlabel('Item Name')
plt.ylabel('Quantity')
plt.title('Top 5 Items by Quantity')
```

```python
items_by_qty = data.groupby('Item Name')['Sub Total'].sum()
items_by_qty.nlargest(5)
```

```python
date=data[data['Date'] == '2023-09-14']
invoice = date.groupby('Invoice No.')['Final Total'].sum().nlargest(1)
invoice
```

```python
total= data.groupby('Invoice No.')['Final Total'].sum()
avg=data['Final Total'].sum()/data['Invoice No.'].nunique()
invo_abv_avg = total[total > avg].index.unique()
print("No of Order's greather than avg bill: ",invo_abv_avg.nunique())
```
